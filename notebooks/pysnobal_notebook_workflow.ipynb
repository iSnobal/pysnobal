This notebook demonstrates how to run pysnobal from a Jupyter Notebook.

```python
import pysnobal.pysnobal as pysnobal
import pandas as pd
```

## Prepare the model configuration dictionary
This reads the model configuration details in config.yaml into a dictionary. You can optionally create the dictionary manually.

```python
config = pysnobal.load_config('../config/config.yaml')
config
```

## Read in forcing data

A separate notebook discusses preparing the input data for pysnobal. Here, you can read in the data specified in config.yaml or provide a path.

```python
forcing_data_df = pd.read_csv('', parse_dates=True, index_col=0)
forcing_data_df
```

## Run the model

```python
output_df = pysnobal.run_snobal(forcing_data_df, config, show_pbar=False)
```

## Inspect the output

```python
output_df['thickness_snow_m'].plot()
```
